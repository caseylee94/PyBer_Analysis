```python
import matplotlib.pyplot as plt
```

```python
# Set the x-asi to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

#Set the y-axis to a list of the floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create a plot with ax.plt()

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create a plot and add a label for the legend
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create a plot and add a label for the legend
plt.plot(x_axis, y_axis, marker = "*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

#Add a grid
plt.grid()

# Add the legend
plt.legend()
```

```python
# Create a plot and add a label for the legend
plt.plot(x_axis, y_axis, marker = ">", color="green", linewidth=2, label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

#Add a grid
plt.grid()

# Add the legend
plt.legend()
```

```python
# Set the x-asi to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

#Set the y-axis to a list of the floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color="green", label='Boston')

#Create labels for axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Create title
plt.title("PyBer Fare by Moth")

#Add legend
plt.legend()
```

```python
# Creaet horizontal bar chart

plt.barh(x_axis, y_axis, color="magenta")
plt.gca().invert_yaxis()
```

```python
# cretae the plot with ax.plt()

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")

#Edit the plot to look better
plt.gca().invert_yaxis
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
```

```python
# Create a scatter plot

plt.plot(x_axis, y_axis, 'o')
```

```python
#Create scatter plot

plt.scatter(y_axis, x_axis, color="red", label='Chicago')
plt.xlabel("Month")
plt.ylabel("Fare($)")

plt.gca().invert_xaxis
plt.title("PyBer Fare by Month")
```

```python
# Bubble chart

#Make data larger to better visualize bubbles

y_axis_larger = []

for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# BUbble chart continued

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Object oriented approach

fig, ax = plt.subplots()
plt.scatter(x_axis, y_axis, s = [i * 5 for i in y_axis], color="skyblue", edgecolors="black", alpha=0.8)
```

```python
# Create pie chart

plt.pie(y_axis, labels=x_axis)

plt.show()
```

```python
#More pie chart stuff

plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')

plt.show()
```

```python
# assing 12 colors one for each piece
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')

plt.show()
```
